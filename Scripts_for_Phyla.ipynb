```python
import os
# print(os.listdir())
import csv
import pandas as pd
from pandas import *
from collections import Counter
import matplotlib.pyplot as plt
import re

def gene_list(name, name1):
    '''Make list of all genes in Hodgkinia genome.'''
    
    filename_open = open(name, 'r')
    data = filename_open.read()
    # print(data)
    
    # read whole genome file without coding sequences annotated
    filename_open1 = open(name1, 'r')
    data1 = filename_open1.read()
    # print(data)
    
    gc_count = 0
    len_count = 0
    data1_split = data1.splitlines()
    for x in range(0, len(data1_split)):
        if data1_split[x].isupper() == True:
            gc_count += data1_split[x].count('G') + data1_split[x].count('C')
            len_count += len(data1_split[x])

    total_gc = gc_count / len_count
    
    gene_name_list = []
    location_list = []
    sequence_list = []
    sequence_gc = []
    
    gene_name_list_1 = []
    location_list_1 = []
    sequence_list_1 = []
    sequence_gc = []
    
    one_gene_info_1 = []
    one_gene_info = []
    
    clean_gene_name = []
    test_list = []
    list_of_lists = []
    
    full_gene_annotation = data.split('>')
    for x in range(1, len(full_gene_annotation)):
        one_gene_info = full_gene_annotation[x].split(']')
        for x in range(0, len(one_gene_info)):
            test_list = []
            if 'gene=' in one_gene_info[x]:
                clean_gene_name = one_gene_info[x].split('[')
                for z in range(0, len(clean_gene_name)):
                    if 'gene=' in clean_gene_name[z]:
                        gene_name_list_1.append(clean_gene_name[z])
                        test_list.append(clean_gene_name[z])
                for y in range(0, len(one_gene_info)): 
                    if 'location=' in one_gene_info[y]:
                        location_list_1.append(one_gene_info[y])
                        test_list.append(one_gene_info[y])
                    if '=' not in one_gene_info[y]:
                        if '|' not in one_gene_info[y]:
                            if 'G' in one_gene_info[y]:
                                if 'C' in one_gene_info[y]:
                                    if one_gene_info[y] != '':
                                        sequence_list_1.append(one_gene_info[y])
                                        test_list.append(one_gene_info[y])
                                        # print(len(test_list))
                                        list_of_lists.append(test_list)
        else:
            if 'gene=' not in one_gene_info[x]:
                gene_name_list.append('unlabeled')
                for y in range(0, len(one_gene_info)): 
                    if 'location=' in one_gene_info[y]:
                        location_list.append(one_gene_info[y])
                    if '=' not in one_gene_info[y]:
                        if '|' not in one_gene_info[y]:
                            sequence_list.append(one_gene_info[y])
    
    gene_name_list_final = []
    location_list_final = []
    sequence_list_final = []
    
    for x in range(0, len(list_of_lists)):
        gene_name_list_final.append(list_of_lists[x][0])
        location_list_final.append(list_of_lists[x][1])
        sequence_list_final.append(list_of_lists[x][2])
    
    for x in range(0, len(sequence_list_1)):
        gc_raw = sequence_list_1[x].count('G') + sequence_list_1[x].count('C')
        gc_content = gc_raw/len(sequence_list_1[x])
        sequence_gc.append(gc_content)
                 
    d = {'Gene name': gene_name_list_final, 'Coordinates': location_list_final, 'Sequence': sequence_list_final, 'GC Content': sequence_gc}
    df = pd.DataFrame(data = d)
    return total_gc
```
